재생목록 곡들 크롤링

```python
import time
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

title = []
song_id = []
rank_by_year = [50927, 43963, 36737, 30871, 25698, 20440, 15389, 11152, 7479, 6683, 6684, 6685, 6709, 6710, 6712, 6713, 6715, 6716, 6717, 6718, 6719, 6708, 6790, 6791, 6792, 6793, 6794, 6795, 6796, 6798]
count = 0
for i in range(30):
    for j in range(i+1, 30):
        if rank_by_year[i] == rank_by_year[j]:
            count += 1
print(count)

for i in rank_by_year:
    driver.get("https://music.bugs.co.kr/musicpd/albumview/" + str(i))
    info_temp = driver.find_elements(By.NAME, "check")
    for k in info_temp:
        song_id.append(k.get_attribute("value"))
        title.append(k.get_attribute("title"))

print(len(title), len(song_id))
```

```python
import time

LYRIC = []
count = 0

for i in song_id:
    print(str(i))
    driver.get("https://music.bugs.co.kr/track/" + str(i) + "?wl_ref=list_tr_08_mab")
    lyric = driver.find_elements(By.TAG_NAME, "xmp")
    count += 1
    print(count)
    for j in lyric:
        LYRIC.append(j.text)

print(len(LYRIC))
df = pd.DataFrame({"제목" : title, "가사" : LYRIC})

df.to_excel(excel_writer="crawling.xlsx", index=True)
```
