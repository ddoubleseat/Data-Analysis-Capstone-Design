```python
import time
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

title = []
song_id = []
playlists = [422618235, 420573216, 467106196, 431011550, 450044182, 501949898, 501153539, 497800568, 500744996, 433029952, 491628537, 496411434]

for i in playlists:
    driver.get("https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq=" + str(i))
    print(i)
    title_temp = driver.find_elements(By.CLASS_NAME, "ellipsis.rank01")
    song_id_temp = driver.find_elements(By.CLASS_NAME, "button_etc.like")
    for j in song_id_temp:
        if j.get_attribute('data-song-no') != None:
            song_id.append(j.get_attribute('data-song-no'))
    for k in title_temp:
        title.append(k.text)
        

print(len(title), len(song_id))
```

```python
import time

LYRIC = []
count = 0

for i in song_id:
    print(str(i))
    driver.get("https://www.melon.com/song/detail.htm?songId=" + str(i))
    lyric = driver.find_elements(By.CLASS_NAME, "lyric")
    count += 1
    print(count)
    for j in lyric:
        if j.text == "":
            LYRIC.append("None")
        else:
            LYRIC.append(j.text)
            print(j.text)

print(len(LYRIC))
df = pd.DataFrame({"제목" : title, "가사" : LYRIC})

df.to_excel(excel_writer="crawling.xlsx", index=True)
```
