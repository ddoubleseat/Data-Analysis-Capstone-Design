```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
# 각자 주소 입력.
wd = webdriver.Chrome("C:\\Users\\MY\\.jupyter\\chromedriver_win32\\chromedriver.exe")
```

```python
from bs4 import BeautifulSoup
import time
import requests
import re
import pandas as pd
```

```python
# 플레이리스트 주소 입력
wd.get('https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq=507114264')
time.sleep(2)
```

```python
song_list = wd.find_elements(By.CSS_SELECTOR, ".btn.button_icons.type03.song_info")
```

```python
len(song_list)
```

```python
title = []
lyric = []
error_count = 0
for i in range(0, len(song_list)):
    song_list = wd.find_elements(By.CSS_SELECTOR,".btn.button_icons.type03.song_info")
    song_list[i].click()
    time.sleep(1)
    
    try:
        t = wd.find_element(By.CSS_SELECTOR,".song_name")
        title.append(t.text)
        time.sleep(1)

        wd.find_element(By.CSS_SELECTOR,'.button_more.arrow_d').click()
        l = wd.find_element(By.CSS_SELECTOR,".lyric.on")
        lyric.append(l.text)
        time.sleep(1)
    
    except:
        error_count += 1
        print("에러 발생 수 : {}".format(error_count))
        print("인덱스 : {}".format(i))
        wd.back()        
        continue
    
    wd.back()
```

```python
result= {"제목" : title , "가사":lyric}  
df = pd.DataFrame(result)  #df로 변환
df
```

```python
# csv 파일로 저장
df.to_csv('result.csv')
```
