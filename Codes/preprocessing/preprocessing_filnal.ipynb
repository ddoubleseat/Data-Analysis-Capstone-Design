```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv("../dataset/result_final.csv", encoding = "utf-8")
```

```python
print(df)
```

```python
df['가사'] = df['가사'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]"," ",regex=True)
```

```python
df['가사'] = df['가사'].str.replace('^ +', "",regex=True) # 공백은 empty 값으로 변경
df['가사'].replace('', np.nan, inplace=True) # 공백은 Null 값으로 변경
df = df.dropna(how='any') # Null 값 제거
```

```python
tokenized_kkma = []

for lyric in tqdm(df.loc[:]["가사"]):
    token = kkma.pos(lyric)
    re_token = [morph[0] for morph in token if len(morph[0]) != 1]
    tokenized_kkma.append(re_token)
```
