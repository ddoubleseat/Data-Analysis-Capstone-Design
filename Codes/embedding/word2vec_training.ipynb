```python
dir = "/content/gdrive/My Drive/Colab Notebooks/" # 노래가사 csv파일 경로
```

```python
df = pd.read_csv(dir + "result_final.csv", encoding = "utf-8")
```

```python
df['가사'] = df['가사'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]"," ",regex=True)
```

```python
df['가사'] = df['가사'].str.replace('^ +', "",regex=True) # 공백은 empty 값으로 변경
df['가사'].replace('', np.nan, inplace=True) # 공백은 Null 값으로 변경
df = df.dropna(how='any') # Null 값 제거
```

```python
print(df)
```

```python
tagging_kkma = []

for i in df.loc[:]["가사"]:
    token = tokenizer.morphs(i)
    tagging_kkma.append(token)
```

```python
pos_kkma = []
for j in tagging_kkma:
    token_word = tokenizer.pos(i)
    pos_kkma.append(token_word)
```

```python
usingwords = []
usingpos = ['NNG', 'VA']
for i in range(len(pos_kkma)):
    for j in range(len(pos_kkma[i])):
            if pos_kkma[i][j][1] in usingpos:
                 usingwords.append(pos_kkma[i][j][0])
```

```python
print(usingwords)
```

```python
print(len(usingwords))
```

```python
usingwords2 = set(usingwords) #집합set으로 변환
usingwords3 = list(usingwords2)
```

```python
print(len(usingwords3))
print(usingwords3)
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(tagging_kkma,
                 size = 100,
                 sg = 0,
                 window = 5,
                 iter = 10,
                 min_count = 1)
```

```python
model.init_sims(replace=True)
```

```python
df1 = pd.DataFrame(model.wv.most_similar("무심", topn = 40), columns = ['단어', '유사도'])
```
